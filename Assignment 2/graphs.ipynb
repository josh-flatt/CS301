```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv", header=None)
data.columns = ["Algorithm", "n", "Runtime"]
data.sort_values(by=["Algorithm"])
data.shape
```

## Graphs
Below, you can see 4 different buckets that algorithms have been placed into. The upper left are quadratic algorithms, the upper right are linear, and the bottom left is constant time. It is important to note that the dictionary functions run in constant time.

    Warning: the dataset for this graph has nearly 1.5 million rows of data. If the following cell is executed, expect a delay beyond one minute for the graph to complete. It likely did not break, it is just taking a very long time.

```python
f, ax = plt.subplots(1, 2, sharey=False, figsize=(20, 10))
sns.lineplot(
    data=data,
    ax=ax[0],
    x=data.n,
    y=data.Runtime,
    hue="Algorithm",
).set(title="Dictionary Methods", ylim=(150, 1550), xlim=(0, 250), xlabel=None, ylabel=None)
sns.lineplot(
    data=data.loc[(data["Algorithm"] == "dict_add") | (data["Algorithm"] == "dict_copy")],
    ax=ax[1],
    x=data.n,
    y=data.Runtime,
    hue="Algorithm",
).set(title="Peculiar behaviors", ylim=(150, 1150), xlim=(0, 250), xlabel=None, ylabel=None);
```
